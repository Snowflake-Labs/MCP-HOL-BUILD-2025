# View pre-evaluated results for this example

In the case you run into trouble performing the evaluation, you can still view the results stored in the database `pre_evaluated.sqlite`.

```python
from trulens.core.session import TruSession
from trulens.core.database.connector.default import DefaultDBConnector


connector = DefaultDBConnector(database_url="sqlite:///pre_evaluated.sqlite")

# Create TruSession with the custom connector
session = TruSession(connector=connector)
```

```python
from trulens.dashboard import run_dashboard
run_dashboard(port=8004)
```
